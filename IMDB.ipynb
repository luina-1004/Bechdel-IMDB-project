```python
# 데이터 변수 리스트 생성
titles = []
years = []
imdb_ratings =[]
metascores = []
votes = []
us_gross = []
```

```python
pages = np.arange(1,13230630,250)
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36", "Accept-Language": "en-US,en"}

for page in pages:
  page = requests.get("https://www.imdb.com/search/title/?release_date=1800-01-01,2023-12-31&count=250&start="+str(page)+"&ref_=adv_prv", headers=headers)
  soup = BeautifulSoup(page.text, 'html.parser')
  movie_div = soup.find_all('div', class_='lister-item mode-advanced')
  #크롤링 속도 제어
  sleep(randint(2,10))
  for con in movie_div:
    #titles
    name = con.h3.a.text
    titles.append(name)
    # years
    year = con.h3.find('span', class_ = 'lister-item-year').text
    years.append(year)
    #imdb
    imdb = con.find('strong').text if con.find('strong') else 0
    imdb_ratings.append(imdb)
    # meta score
    meta = con.find('span', class_='metascore').text if con.find('span', class_='metascore') else ''
    # if len(con.find('span', class_='metascore').text) > 1:
    #   meta = con.find('span', class_='metascore').text
    metascores.append(meta)
    # else: metascores.append('')
    # votes
    # vote = con.find('span', attrs={'name':"nv"})['data-value'] # attrs: 속성 값 모두 출력
    # if len(vote)>1:
    #   votes.append(int(vote))
    # else:
    #   votes.append('')
    vote = con.find_all('span', attrs={'name':"nv"})[0].text if con.find_all('span', attrs={'name':"nv"}) else 0 
    votes.append(vote)
    # us_gross
    gross = con.find_all('span', attrs={'name':'nv'})[1].text if len(con.find_all('span', attrs={'name':'nv'}))>1 else ''
    us_gross.append(gross)

  print("Done")
```

```python
movies = pd.DataFrame({
    'movie' : titles,
    'year' : years,
    'imdb' : imdb_ratings,
    'metascore': metascores,
    'votes' : votes,
    'us_gross_dollars': us_gross 
})
  
```
