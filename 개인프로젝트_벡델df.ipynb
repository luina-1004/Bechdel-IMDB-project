```python
# 데이터 변수 리스트 생성
titles = []
years = []
imdb_ratings =[]
metascores = []
votes = []
us_gross = []
```

```python
movie_div = soup.find_all('div', class_ = 'lister-item mode-advanced')
```

```python
# titles
for con in movie_div:
  name = con.h3.a.text
  titles.append(name)
```

```python
titles
```

```python
# years
for con in movie_div:
  year = con.h3.find('span', class_ = 'lister-item-year').text
  years.append(year)
```

```python
years
```

```python
# imdb rating
for con in movie_div:
  imdb = float(con.find('strong').text)
  imdb_ratings.append(imdb)
```

```python
# meta score
for con in movie_div:
  meta = con.find('span', class_='metascore').text
  if len(meta) > 1:
    metascores.append(meta.strip())
  else: metascores.append(np.nan)
```

```python
#https://truman.tistory.com/107
#votes = []
for con in movie_div:
  vote = con.find('span', attrs={'name':"nv"}).text#['data-value'] # attrs: 속성 값 모두 출력
  # vote = con.find_all('span', attrs={'name':"nv"})[0].text # attrs: 속성 값 모두 출력

  print(vote)
  # if len(vote)>1:
    # votes.append(int(vote))
  # else:
    # votes.append(np.nan)
```

```python
votes
```

```python
# us_gross = []
for con in movie_div:
  gross = con.find('span', attrs={'name':'nv'}).findNext('span', attrs={'name':'nv'})['data-value'].replace(',','')
  if len(gross) > 3:
    us_gross.append(int(gross))
  else: us_gross.append(np.nan)
```

```python
us_gross
```

```python
df = pd.DataFrame({
    'movie' : titles,
    'year' : years,
    'imdb' : imdb_ratings,
    'metascore': metascores,
    'votes' : votes,
    'us_gross_dollars': us_gross 
})
  
```
