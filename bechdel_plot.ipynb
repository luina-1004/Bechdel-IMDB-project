```python
year = list(bechdel['year'].unique())
```

```python
# year - rating(mean)
movie = []
for y in year:
    # movie = []
    # movie.append(bechdel[bechdel['year'] == y])
    movie.append(bechdel[bechdel['year'] == y])

```

```python
mean = []
for i in range(len(movie)):
    mean.append(movie[i]['rating'].mean().round(2))

print(mean)
```

```python
plt.figure()
plt.plot(year,mean)
plt.xlabel('Year')
plt.ylabel('Average Bechdel Rating')
plt.title('Average Bechdel Rate via Movie Year')

# plt.savefig('year-bechdel.png', dpi=400, bbox_inches='tight')
plt.show()
```
